```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables
N = [1, 2, 4, 8, 10, 100, 1000]
cos_sim_path = [0.04, 0.0599, 0.0595, 0.064, 0.066, 0.1318, 0.27]
cos_sim_std_path = [0.0043, 0.00425, 0.00463, 0.0051, 0.0052, 0.0052, 0.0062]
cos_sim_rein = [-0.0047, 0.00153, 0.0075, 0.0085, 0.011, 0.039, 0.12]
cos_sim_std_rein = [0.0032, 0.0029, 0.003, 0.003, 0.002, 0.0032, 0.0026]

# Set new x-axis values (1, 2, 3, 4, 5, 6, 7 to match the original data length)
x_values = [1, 2, 3, 4, 5, 6, 7]

# Calculate adjusted standard errors
adjustment_factor = 1.96
adjusted_cos_sim_std_path = [std * adjustment_factor for std in cos_sim_std_path]
adjusted_cos_sim_std_rein = [std * adjustment_factor for std in cos_sim_std_rein]

# Set up the plot
plt.figure(figsize=(14, 10))

# Plot mean values with adjusted error bars for standard errors
plt.errorbar(x_values, cos_sim_path, yerr=adjusted_cos_sim_std_path, fmt='-o', label='Autodiff', capsize=5)
plt.errorbar(x_values, cos_sim_rein, yerr=adjusted_cos_sim_std_rein, fmt='-o', label='REINFORCE', capsize=5)

# Set x-axis and y-axis ticks with larger font size
plt.xticks(x_values,[str(_) for _ in N] ,fontsize=35)
plt.yticks(fontsize=35)

# Set labels with large text
plt.xlabel('$N$', fontsize=35)
plt.ylabel('Cosine Similarity', fontsize=35)
#plt.title('Cosine Similarity with Adjusted Standard Error', fontsize=35)
plt.legend(fontsize=35)


plt.savefig('compare_cos.pdf', format='pdf', bbox_inches='tight')
# Show the plot without grid
plt.show()
```
