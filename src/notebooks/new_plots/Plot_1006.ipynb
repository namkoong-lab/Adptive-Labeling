```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables
N = [1, 2, 4, 8, 10, 100, 1000]
cos_sim_path = [0.04, 0.0599, 0.0595, 0.064, 0.066, 0.1318, 0.27]
cos_sim_std_path = [0.0043, 0.00425, 0.00463, 0.0051, 0.0052, 0.0052, 0.0062]
cos_sim_rein = [-0.0047, 0.00153, 0.0075, 0.0085, 0.011, 0.039, 0.12]
cos_sim_std_rein = [0.0032, 0.0029, 0.003, 0.003, 0.002, 0.0032, 0.0026]

# Set new x-axis values (1, 2, 3, 4, 5, 6, 7 to match the original data length)
x_values = [1, 2, 3, 4, 5, 6, 7]

# Calculate adjusted standard errors
adjustment_factor = 1.96
adjusted_cos_sim_std_path = [std * adjustment_factor for std in cos_sim_std_path]
adjusted_cos_sim_std_rein = [std * adjustment_factor for std in cos_sim_std_rein]

# Set up the plot
plt.figure(figsize=(14, 10))

# Plot mean values with adjusted error bars for standard errors
plt.errorbar(x_values, cos_sim_path, yerr=adjusted_cos_sim_std_path, fmt='-o', label='Autodiff', capsize=5)
plt.errorbar(x_values, cos_sim_rein, yerr=adjusted_cos_sim_std_rein, fmt='-o', label='REINFORCE', capsize=5)

# Set x-axis and y-axis ticks with larger font size
plt.xticks(x_values,[str(_) for _ in N] ,fontsize=35)
plt.yticks(fontsize=35)

# Set labels with large text
plt.xlabel('$N$', fontsize=35)
plt.ylabel('Cosine Similarity', fontsize=35)
#plt.title('Cosine Similarity with Adjusted Standard Error', fontsize=35)
#plt.legend(fontsize=35)


plt.savefig('compare_grad_cos.pdf', format='pdf', bbox_inches='tight')
# Show the plot without grid
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables for MSE
N = [1, 2, 4, 8, 10, 100, 1000]
mse_avg_pathwise = [6.5e-6, 3.5e-6, 1e-6, 4.6e-7, 3.2e-7, 2.37e-8, 2.9e-9]
mse_std_error_pathwise = [5.7e-6, 1.9e-6, 4.9e-7, 1.9e-7, 1.25e-7, 5.2e-9, 2.8e-10]
mse_avg_reinf = [1.2e-5, 6.3e-6, 3.2e-6, 1.6e-6, 1.3e-6, 1.29e-7, 1.27e-8]
mse_std_error_reinf = [3.4e-7, 1.26e-7, 4.5e-8, 1.96e-8, 1.36e-8, 7.02e-10, 5.7e-11]

# Set new x-axis values (1, 2, 3, 4, 5, 6, 7 to match the original data length)
x_values = [1, 2, 3, 4, 5, 6, 7]

# Calculate adjusted standard errors (1.96 is typically used for a 95% confidence interval)
adjustment_factor = 1.96
adjusted_mse_std_error_pathwise = [std * adjustment_factor for std in mse_std_error_pathwise]
adjusted_mse_std_error_reinf = [std * adjustment_factor for std in mse_std_error_reinf]

# Set up the plot
plt.figure(figsize=(14, 10))

# Plot mean values with adjusted error bars for standard errors
plt.errorbar(x_values, mse_avg_pathwise, yerr=adjusted_mse_std_error_pathwise, fmt='-o', label='Autodiff', capsize=5)
plt.errorbar(x_values, mse_avg_reinf, yerr=adjusted_mse_std_error_reinf, fmt='-o', label='REINFORCE', capsize=5)

# Set x-axis and y-axis ticks with larger font size
plt.xticks(x_values, [str(_) for _ in N], fontsize=35)
plt.yticks(fontsize=35)

# Set labels with large text
plt.xlabel('$N$', fontsize=35)
plt.ylabel('Mean Squared Error', fontsize=35)

# Add legend with a larger font size
plt.legend(fontsize=35, loc = 'upper left')

# Save the plot as a PDF file with a tight layout
plt.savefig('compare_grad_mse.pdf', format='pdf', bbox_inches='tight')
plt.yscale('log')
# Show the plot without grid
plt.show()
```
